### **Question 9**: What percentage of the orders cost more than 20 dollars? [2 marks]

```python
# ATTENTION: AI-generated code can include errors or operations you didn't intend. Review the code in this cell carefully before running it.

from pyspark.sql import functions as F

# Filter data for orders with a cost greater than 20 dollars
expensive_orders = df.filter(df["cost_of_the_order"] > 20)

# Calculate the total number of expensive orders
expensive_order_count = expensive_orders.count()

# Calculate the total number of orders
total_order_count = df.count()

# Calculate the percentage of orders that cost more than 20 dollars
percentage_expensive_orders = (expensive_order_count / total_order_count) * 100

print("Percentage of orders costing more than $20: {:.2f}%".format(percentage_expensive_orders))
```

### **Question 11:** The company has decided to give 20% discount vouchers to the top 3 most frequent customers. Find the IDs of these customers and the number of orders they placed. [1 mark]

```python
# ATTENTION: AI-generated code can include errors or operations you didn't intend. Review the code in this cell carefully before running it.

from pyspark.sql import functions as F

# Group by customer ID and count the number of orders placed by each customer
customer_order_counts = df.groupBy("customer_id").count()

# Sort customers in descending order based on order count
top_customers = customer_order_counts.orderBy(F.desc("count"))

# Select the top 3 most frequent customers and their order counts
top_3_customers = top_customers.select("customer_id", "count").limit(3)

# Display the IDs of the top 3 most frequent customers and their order counts
top_3_customers.show()
```

### **Question 14:** The company charges the restaurant 25% on the orders having cost greater than 20 dollars and 15% on the orders having cost greater than 5 dollars. Find the net revenue generated by the company across all orders. [3 marks]

```python


# function to determine the revenue
def compute_rev(x):
    if x > 20:
        return x * 0.25
    elif x > 5:
        return x * 0.15
    else:
        return x

df['Revenue'] = df['cost_of_the_order'].apply(compute_rev)
df.head()
```

# Observation : 
The provided code snippet defines a function compute_rev that calculates revenue based on the input value x. If x is greater than 20, the revenue is calculated as 25% of x. If x is between 6 and 20, the revenue is calculated as 15% of x. If x is less than or equal to 5, the revenue remains the same as x.

The next line applies this function to a column named 'cost_of_the_order' in a DataFrame called df, and stores the computed revenues in a new column named 'Revenue'. This operation is done using the .apply() method in pandas.

Overall, this code segment is used to calculate revenue based on different conditions and store it in a new column within a DataFrame. It provides flexibility in determining revenue based on varying input values.

```python


total_rev = df['Revenue'].sum()  ## Calculate the total revenue by summing the 'Revenue' column
print('The net revenue is around', round(total_rev, 2), 'dollars')  ## Print the total revenue rounded to 2 decimal places
```

# Observation:

The provided code snippet calculates the total revenue by summing the values in the 'Revenue' column of the DataFrame df. It then prints out the total revenue rounded to two decimal places.

The total_rev variable stores the sum of all revenue values in the 'Revenue' column.
The print statement displays a message indicating the total revenue, rounded to two decimal places, providing a concise summary of the overall revenue generated.
This calculation is useful for understanding the total financial impact or income generated based on the revenue calculations performed earlier in the DataFrame.
